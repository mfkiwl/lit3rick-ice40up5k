```python
import numpy as np
import matplotlib.pyplot  as plt

from numpy import genfromtxt

try:
    import cPickle as pickle
except ImportError:  # Python 3.x
    import pickle
```

# Doing first HV pulses at different gains

```python
plt.figure(figsize=(15,5))

d={}
t = [x/64.0 for x in range(64*128)]
d["t"] = t

for hiloVal in [0,1]:
    for DAC in range(9+2):
        DAC = DAC*50
        my_data = genfromtxt('tmpdata.pos_pulse/dac'+str(DAC)+'_'+str(hiloVal)+'.csv', delimiter=';')
        d[hiloVal,DAC] = my_data/1024.0
        plt.plot(t,my_data/1024.0)

with open('lit3rick90V_PositivePulse.pkl', 'wb') as fp:
    pickle.dump(d, fp, protocol=4)
    
plt.title("Lit3rick, positive pulse, 90V, acquisitions")
plt.savefig("lit3_90V_ppulse.jpg")
```

# Injecting HV pulses

```python
plt.figure(figsize=(15,5))

d={}
t = [x/64.0 for x in range(64*128)]
d["t"] = t

for hiloVal in [0,1]:
    for DAC in range(9+2):
        DAC = DAC*50
        my_data = genfromtxt('tmpdata.neg_pulse/dac'+str(DAC)+'_'+str(hiloVal)+'.csv', delimiter=';')
        d[hiloVal,DAC] = my_data/1024.0
        plt.plot(t,my_data/1024.0)

with open('lit3rick90V_NegativePulse.pkl', 'wb') as fp:
    pickle.dump(d, fp, protocol=4)
    
plt.title("Lit3rick, negative pulse, 90V, acquisitions")
plt.savefig("lit3_90V_npulse.jpg")
```

# Playing with the pulse width

```python
plt.figure(figsize=(15,5))

d={}
t = [x/64.0 for x in range(64*128)]
d["t"] = t

for k in range(10):
    k = k*5
        
    my_data = genfromtxt('tmpdata.pulse_width/pulsewidth_'+str(k)+'.csv', delimiter=';')
    d[k] = my_data/1024.0
    plt.plot(t,my_data/1024.0,label="Width = "+str(k) )

with open('lit3rick90V_pulsewidth.pkl', 'wb') as fp:
    pickle.dump(d, fp, protocol=4)

plt.legend()
plt.title("Lit3rick, hilo = 0, gain=100, 90V, different pulse width")
plt.savefig("lit3_90V_width.jpg")
```
