```python
import numpy as np
import matplotlib.pyplot  as plt

from numpy import genfromtxt

try:
    import cPickle as pickle
except ImportError:  # Python 3.x
    import pickle
```

```python
plt.figure(figsize=(15,5))

d={}
t = [x/64.0 for x in range(64*128)]
d["t"] = t

for hiloVal in [0,1]:
    for DAC in range(9+2):
        DAC = DAC*50
        my_data = genfromtxt('tmpdata.base/dac'+str(DAC)+'_'+str(hiloVal)+'.csv', delimiter=';')
        d[hiloVal,DAC] = my_data/1024.0
        plt.plot(t,my_data/1024.0)
my_data = genfromtxt('tmpdata.base/noise.csv', delimiter=';')
d["noise"] = my_data/1024.0

with open('lit3rick.pkl', 'wb') as fp:
    pickle.dump(d, fp, protocol=4)
    
plt.title("Lit3rick, 5V, acquisitions")
plt.savefig("lit3_5V_ppulse.jpg")
```
